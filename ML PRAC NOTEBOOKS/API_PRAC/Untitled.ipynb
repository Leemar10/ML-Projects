```python
import requests
import pandas as pd
```

```python
response = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=b0e9edc35da2527dedf89879dd692549")
```

```python
pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
```

```python
df = pd.DataFrame()
```

```python
for i in range(1, 216):
    response = requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=b0e9edc35da2527dedf89879dd692549&language=en-US&page={i}")
    temp_df = pd.DataFrame(response.json()['results'])
    df = pd.concat([df,temp_df], ignore_index=True)
```

```python
for i in range(216, 512):
    response = requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=b0e9edc35da2527dedf89879dd692549&language=en-US&page={i}")
    temp_df = pd.DataFrame(response.json()['results'])
    df = pd.concat([df,temp_df], ignore_index=True)
```
